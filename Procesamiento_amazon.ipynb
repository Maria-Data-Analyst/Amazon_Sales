```python
#####  Función para limpiar la columna eliminando todos los símbolos excepto números, puntos y comas ####
def clean_column(column):
    # Mantener solo números, puntos y comas, eliminando otros símbolos
    return column.replace('[^0-9.,]', '', regex=True)

# Aplicar la función a las columnas
df_producto['discounted_price'] = clean_column(df_producto['discounted_price'])
df_producto['actual_price'] = clean_column(df_producto['actual_price'])
df_producto['discount_percentage'] = clean_column(df_producto['discount_percentage'])
```

```python
##### Función para reemplazar comas por puntos y eliminar comas usadas como separadores de miles #####
def format_column(column):
    # Asegúrate de que la columna es de tipo cadena (str) para aplicar las transformaciones
    column = column.astype(str)
    # Primero, eliminamos las comas usadas como separadores de miles
    column = column.str.replace(',', '', regex=False)
    # Luego, si es necesario, reemplazamos los puntos por puntos decimales
    # Si tus datos usan puntos como decimales, descomenta la siguiente línea
    # column = column.str.replace('.', '.', regex=False)  # Esta línea se puede ajustar o eliminar si no es necesaria
    return column

# Aplicar la función a las columnas
df_producto['discounted_price'] = format_column(df_producto['discounted_price'])
df_producto['actual_price'] = format_column(df_producto['actual_price'])
df_producto['discount_percentage'] = format_column(df_producto['discount_percentage'])
```

```python
## bases incorrectas en el link de imagen ##

base_url = "https://m.media-amazon.com/images/I"

# Función para convertir a cadena, pero respetar los valores NaN
def convert_to_str(value):
    if pd.isnull(value):
        return value  # Mantener NaN como NaN
    return str(value)

# Convertir todos los valores en 'img_link' a cadenas, respetando los valores NaN
df_review['img_link'] = df_review['img_link'].apply(convert_to_str)

# Filtrar filas donde 'img_link' no es nulo y no comienza con la base URL correcta
non_null_links = df_review[~df_review['img_link'].isnull()]
incorrect_links = non_null_links[~non_null_links['img_link'].str.startswith(base_url)]

# Extraer la parte de la URL antes de "I/"
def extract_before_I(url):
    if pd.isnull(url):  # Verificar si el valor es NaN
        return url  # Mantener NaN como NaN
    # Encuentra el índice de "I/" y extrae todo lo que está antes
    index = url.find("I/")
    if index != -1:
        return url[:index]
    return url

# Aplicar la extracción a las URLs incorrectas
incorrect_links['before_I'] = incorrect_links['img_link'].apply(extract_before_I)

# Obtener URLs únicas de la parte antes de "I/"
unique_before_I = incorrect_links['before_I'].unique()

# Mostrar los resultados
print("Número de bases incorrectas únicas:", len(unique_before_I))
print("Bases incorrectas únicas:")
for base in unique_before_I:
    print(base)
```

```python
## Visualizar otros patrones incorrectos en el enlace imagen ##
def check_image_links(df):
    # Define la parte fija del enlace
    fixed_part = 'https://m.media-amazon.com/images/I/'

    # Función para verificar el orden de los patrones
    def check_order(link):
        if pd.isnull(link):  # Verificar si el valor es NaN
            return None
        if not link.startswith(fixed_part):
            return None
        # Extrae la parte después de 'I/' y antes de '.jpg'
        middle_part = link[len(fixed_part):-len('.jpg')]
        # Busca los patrones en el orden correcto
        if '._SX300_SY300' in middle_part:
            return 'Correct'
        elif '._SY300_SX300' in middle_part:
            return 'Incorrect'
        return 'Other'

    # Aplica la función a cada enlace
    df['order_check'] = df['img_link'].apply(check_order)

    # Identifica los patrones incorrectos únicos
    incorrect_links = df[df['order_check'] == 'Incorrect']

    # Extrae el patrón incorrecto de la parte del enlace
    def extract_pattern(link):
        if pd.isnull(link):
            return None
        parts = link.split('._')
        if len(parts) < 2:
            return None
        return parts[1].split('_')[0] + '_' + parts[1].split('_')[1]

    incorrect_links['pattern'] = incorrect_links['img_link'].apply(extract_pattern)

    # Obtén patrones únicos diferentes a 'SX300_SY300'
    unique_incorrect_patterns = incorrect_links['pattern'].unique()
    unique_incorrect_patterns = [pat for pat in unique_incorrect_patterns if pat != 'SX300_SY300']

    # Filtra los enlaces que contienen estos patrones incorrectos únicos
    filtered_incorrect_links = incorrect_links[incorrect_links['pattern'].isin(unique_incorrect_patterns)]

    # Imprime el conteo y los enlaces con patrones incorrectos únicos
    print(f"Total de patrones incorrectos únicos diferentes a 'SX300_SY300': {len(unique_incorrect_patterns)}")
    print("Patrones incorrectos únicos:")
    print(unique_incorrect_patterns)
    #print("Enlaces con patrones incorrectos únicos:")
    #print(filtered_incorrect_links[['img_link', 'pattern']])

# Llama a la función con tu DataFrame
check_image_links(df_review)

```

```python

# Definir las partes que deben ser reemplazadas en la base del link de imagen
replace_dict = {
    "https://m.media-amazon.com/images/W/WEBP_402378-T1/images/": "https://m.media-amazon.com/images/",
    "https://m.media-amazon.com/images/W/WEBP_402378-T2/images/": "https://m.media-amazon.com/images/",
    "https://images-na.ssl-images-amazon.com/images/W/WEBP_402378-T2/images/": "https://m.media-amazon.com/images/"
}

# Función para reemplazar las partes específicas y mantener el resto de la URL
def replace_url(url):
    if pd.isnull(url):  # Verificar si el valor es NaN
        return url  # Mantener el valor nulo sin cambios
    for old, new in replace_dict.items():
        if url.startswith(old):
            # Reemplazar la parte vieja con la nueva y mantener el resto de la URL
            return new + url[len(old):]
    return url

# Aplicar la función de reemplazo a la columna 'img_link'
df_review['img_link'] = df_review['img_link'].apply(replace_url)

# Mostrar las primeras filas para verificar
print(df_review.head())

```

```python
### Reemplaza los demas errores en la rita de enlace de las iamgenes para dejar funcinando el link
import re
def check_and_correct_image_links(df_review):
    # Define la parte fija del enlace (aunque no se usa en la función actual)
    fixed_part = 'https://m.media-amazon.com/images/I/'

    # Función para corregir los enlaces con patrones incorrectos
    def correct_link(link):
        if pd.isnull(link):  # Verificar si el enlace es NaN
            return link  # Mantener el valor nulo sin cambios
        # Extrae la parte antes de '_SX' o '_SY'
        before_pattern = re.split(r'(_SX\d+|_SY\d+)', link)[0]
        # Combina la parte antes del patrón con el patrón correcto
        return f'{before_pattern}_SX300_SY300.jpg'

    # Aplica la función de corrección a los enlaces
    df_review['img_link'] = df_review['img_link'].apply(correct_link)

    # Muestra los primeros enlaces corregidos para verificar
    print("Enlaces corregidos:")
    print(df_review.head())

# Llama a la función con tu DataFrame
check_and_correct_image_links(df_review)
```
